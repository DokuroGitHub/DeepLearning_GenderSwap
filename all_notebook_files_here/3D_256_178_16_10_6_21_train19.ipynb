#functions

##crop face using cv2

```python
def crop_face(imgarray, section, margin=20, size=256):
    """
    :param imgarray: full image
    :param section: face detected area (x, y, w, h)
    :param margin: add some margin to the face detected area to include a full head
    :param size: the result image resolution with be (size x size)
    :return: resized image in numpy array with shape (size x size x 3)
    """
    img_h, img_w, _ = imgarray.shape
    if section is None:
        section = [0, 0, img_w, img_h]
    (x, y, w, h) = section
    margin = int(min(w, h) * margin / 100)
    x_a = x - margin
    y_a = y - margin
    x_b = x + w + margin
    y_b = y + h + margin
    if x_a < 0:
        x_b = min(x_b - x_a, img_w - 1)
        x_a = 0
    if y_a < 0:
        y_b = min(y_b - y_a, img_h - 1)
        y_a = 0
    if x_b > img_w:
        x_a = max(x_a - (x_b - img_w), 0)
        x_b = img_w
    if y_b > img_h:
        y_a = max(y_a - (y_b - img_h), 0)
        y_b = img_h
    cropped = imgarray[y_a: y_b, x_a: x_b]
    print('cropped:',cropped.shape)
    print('y_a:',y_a,'y_b:',y_b,'x_a:',x_a,'x_b:',x_b)
    resized_img = cv2.resize(cropped, (size, size), interpolation=cv2.INTER_AREA)
    resized_img = np.array(resized_img)
    print('resized_img:',resized_img.shape)
    return resized_img,y_a,y_b,x_a,x_b
```

##normalize img

```python
def normalize(arr):
    """
    Linear normalization
    http://en.wikipedia.org/wiki/Normalization_%28image_processing%29
    """
    arr = arr.astype('float')
    # Do not touch the alpha channel
    for i in range(3):
        minval = arr[...,i].min()
        maxval = arr[...,i].max()
        if minval != maxval:
            arr[...,i] -= minval
            arr[...,i] *= (255.0/(maxval-minval))
    return arr
```

##pre_processed_img(imgFile)

```python
def pre_processed_image(imgFile):
    full_path = test_dir+imgFile
    print(full_path)
    frame = cv2.imread(full_path)
    if (frame is not None):
        frame = frame[:, :, ::-1] # CV2'BGR -> RGB
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(
            gray,
            scaleFactor=1.2,
            minNeighbors=10
        )
        if len(faces) is 0:
            print('No face detected, try another image.')
            return
        main_face = max(faces, key=lambda rectangle: (rectangle[2] * rectangle[3]))  # area = w * h
        face_img,y_a,y_b,x_a,x_b = crop_face(frame, main_face, margin = 40, size=256)
        print('main_face:',main_face.shape,main_face)
        print('face_img:',face_img.shape)
        image = Image.fromarray(face_img)

        # Pre-process the image
        transform = []
        transform.append(T.ToTensor())
        transform.append(T.Normalize(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5)))
        transform = T.Compose(transform)
        preprocessed_image = transform(face_img)
        print('preprocessed_image,y_a,y_b,x_a,x_b:',preprocessed_image.shape,y_a,y_b,x_a,x_b)
        return preprocessed_image,y_a,y_b,x_a,x_b
```

## Generator

```python
#from google.colab import files
import cv2
from google.colab.patches import cv2_imshow
#plt.savefig("abc.png")
#test_dir = '/content/drive/MyDrive/GROUP05_FINAL_ALL_DATA/3d_256_178_16/inputs/'
#results_dir =  '/content/drive/MyDrive/GROUP05_FINAL_ALL_DATA/3d_256_178_16/results'
def tryCvStarGAN(imgFile, setLipMakeUp = False,setNoBeard=False, setMale = False):
    if (pre_processed_image(imgFile) is not None):
        preprocessed_image,y_a,y_b,x_a,x_b = pre_processed_image(imgFile)
        # Show the pre-processed source image.
        # Prepare the plot
        fig=plt.figure(figsize=(8, 8))
        columns = 2
        rows = 1
        fig.add_subplot(rows, columns, 1)
        preprocessed_image_to_show1 =np.moveaxis(preprocessed_image.numpy(),[0], [2])
        plt.imshow((preprocessed_image_to_show1+1)/2)
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        # Pre-process the image
        
        labels = torch.zeros(c_dim)
        # Set the 'LipMakeUp' label.
        if setLipMakeUp is True:
            labels[0] = 1
        # Set the 'Beard' label.
        if setNoBeard is True:
            labels[1] = 1
        # Set the 'Male' label.
        if setMale is True:
            labels[2] = 1
        # Run the generator to generate the desired image with labels.
        print(preprocessed_image.unsqueeze(0).to(device).shape)
        #x_fake = G(x_real, c_trg)
        print(labels)
        generated = G(preprocessed_image.unsqueeze(0).to(device), labels.unsqueeze(0).to(device))
        print('generated',generated.shape)
        #out_src, out_cls = D(x_fake)
        out_src, out_cls = D(preprocessed_image.unsqueeze(0).to(device))
        print(out_cls)
        out_src2, out_cls2 = D(generated)
        print(out_cls2)

        # Show the generated image.
        fig.add_subplot(rows, columns, 2)
        generated_img_256x256 = np.moveaxis(generated.cpu().detach().numpy()[0],[0], [2])
        print('generated_img_256x256',generated_img_256x256.shape)
        #generated_img_to_fill = cv2.resize(generated_img_256x256, (315, 315), interpolation=cv2.INTER_AREA)
        generated_img_to_fill = cv2.resize(generated_img_256x256, (x_b-x_a, y_b-y_a), interpolation=cv2.INTER_AREA)
        generated_img_to_fill = normalize(generated_img_to_fill).astype('uint8')
        generated_img_to_fill = generated_img_to_fill[:, :, ::-1] # CV2'BGR -> RGB
        #origin image
        org_frame = cv2.imread(test_dir+imgFile)
        print('org_frame',org_frame.shape)
        print('generated_img_to_fill',generated_img_to_fill.shape)
        org_frame[y_a: y_b, x_a: x_b] = generated_img_to_fill
        #plt.imshow((preprocessed_image_to_show1+1)/2)
        plt.imshow((generated_img_256x256+1)/2)
        #save image
        plt.savefig(f"{results_dir}/before_and_after/{imgFile}")
        print(f"{results_dir}/before_and_after/{imgFile} saved")
        plt.show()

        #save_full_size_as_org_img
        cv2.imwrite(f"{results_dir}/full_size/{imgFile}", org_frame)
        print(f"{results_dir}/full_size/{imgFile} saved")
        cv2_imshow(org_frame)
```

## to Female

```python
def swapToFemale(imgPath):
    if (pre_processed_image(imgPath) is not None):
        preprocessed_image,y_a,y_b,x_a,x_b = pre_processed_image(imgPath)
        minScore = 1000
        for isLipMakeUp in range(0,2):
            for isNoBeard in range(0,2):
                labels = torch.zeros(c_dim)  #0 0 0
                labels[0] = isLipMakeUp
                labels[1] = isNoBeard
                labels[2] = 0
                generated = G(preprocessed_image.unsqueeze(0).to(device), labels.unsqueeze(0).to(device))
                out_src, out_cls = D(generated)
                maleScore = out_cls[0][2].item()
                if (maleScore<minScore):
                    minScore = maleScore
                    if (isLipMakeUp==1):
                        setLipMakeUp = True
                    else:
                        setLipMakeUp = False
                    if (isNoBeard==1):
                        setNoBeard = True
                    else:
                        setNoBeard = False
        print('minScore= ',minScore)
        print('Swapping to female... set: setLipMakeUp=',setLipMakeUp, ' setNoBeard= ',setNoBeard)
        tryCvStarGAN(imgPath, setLipMakeUp = setLipMakeUp,setNoBeard=setNoBeard, setMale = False)
```

## to Male

```python
def swapToMale(imgPath):
    if (pre_processed_image(imgPath) is not None):
        preprocessed_image,y_a,y_b,x_a,x_b = pre_processed_image(imgPath)
        maxScore = -1000
        for isLipMakeUp in range(0,2):
            for isNoBeard in range(0,2):
                labels = torch.zeros(c_dim)  #0 0 0
                labels[0] = isLipMakeUp
                labels[1] = isNoBeard
                labels[2] = 0
                generated = G(preprocessed_image.unsqueeze(0).to(device), labels.unsqueeze(0).to(device))
                out_src, out_cls = D(generated)
                maleScore = out_cls[0][2].item()
                if (maleScore>maxScore):
                    minScore = maleScore
                    if (isLipMakeUp==1):
                        setLipMakeUp = True
                    else:
                        setLipMakeUp = False
                    if (isNoBeard==1):
                        setNoBeard = True
                    else:
                        setNoBeard = False
        print('maxScore= ',maxScore)
        print('Swapping to male... set: setLipMakeUp=',setLipMakeUp, ' setNoBeard= ',setNoBeard)
        tryCvStarGAN(imgPath, setLipMakeUp = setLipMakeUp,setNoBeard=setNoBeard, setMale = True)
```

## swap

```python
def swapGender(imgPath):
    if (pre_processed_image(imgPath) is not None):
        preprocessed_image,y_a,y_b,x_a,x_b = pre_processed_image(imgPath)
        out_src, out_cls = D(preprocessed_image.unsqueeze(0).to(device))
        print("out_src:", out_src.shape,out_src)
        print("out_cls:", out_cls)
        maleScore = out_cls[0][2].item()
        print(maleScore)
        if (maleScore > 0 ):
            print('Male detected with maleScore= ',maleScore)
            swapToFemale(imgPath)
        else:
            print('Female detected with maleScore= ',maleScore)
            swapToMale(imgPath)
    else:
        print("can not detect")
```
